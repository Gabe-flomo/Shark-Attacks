```python
# classifying the activity as one of these 5 classes

df["swimming"] = df["activity"].str.contains("swim",case = False,regex = False)
df["fishing"] = df["activity"].str.contains("fish",case = False,regex = False)
df["surfing"] = (df["activity"].str.contains("surf",case = False,regex = False)) | (df["activity"].str.contains("paddl",case = False,regex = False)) | (df["activity"].str.contains("board",case = False,regex = False))
df["diving"] = df["activity"].str.contains("diving",case = False,regex = False)
df["sailing"] = (df["activity"].str.contains("boat",case = False,regex = False)) | (df["activity"].str.contains("capsize",case = False,regex = False)) | (df["activity"].str.contains("ship",case = False,regex = False))
df["other"] = (swimming != True) & (fishing != True) & (surfing != True) & (diving != True) & (sailing != True)
```
